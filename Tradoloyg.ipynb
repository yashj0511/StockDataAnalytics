```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("/content/tradelog.csv")
column_names = ['ID','Ticker','Entry Time','Entry Price','Exit Time','Exit Price']
df.columns = column_names
df.head()
df['Ticker'].unique()
print(df['Ticker'].nunique()) #unique ticker values count
print(df.shape[0]) #total rows in dataframe
```

```python
# for i in df:
#   print(i)
#   # print(i['Entry Price'])
for index, row in df.iterrows():
    print(f" Name: {row['Ticker']}, Entry Time: {row['Entry Time']}, Exit Time: {row['Exit Time']}, Entry Price:{row['Entry Price']},Exit Price:{row['Exit Price']}")
```

```python
Total_Trades_count=len(df)

print("Total Trades ",Total_Trades_count)
profitable_trades=[]
loss_trades=[]
for index, row in df.iterrows():
    if row['Entry Price']<=row['Exit Price']:
      profitable_trades.append(row)
    else:
      loss_trades.append(row)

# profitable_trades
df_ProfitableTrade=pd.DataFrame(profitable_trades, columns=['ID','Ticker','Entry Time','Entry Price','Exit Time','Exit Price'])
df_ProfitableTrade

#Loss trades
df_LossTrade=pd.DataFrame(loss_trades, columns=['ID','Ticker','Entry Time','Entry Price','Exit Time','Exit Price'])


#count
profitable_trades_count=len(df_ProfitableTrade)
print("Profit makinng trades ",profitable_trades_count)

loss_trades_count=len(df_LossTrade)

print("Loss Making Trades ",loss_trades_count)

```

```python
win_rate=profitable_trades_count/Total_Trades_count
print("Win rate ",win_rate)
```
